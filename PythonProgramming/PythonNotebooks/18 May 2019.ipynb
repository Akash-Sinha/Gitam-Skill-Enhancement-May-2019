### Problem 3 : 
#### Problem Statement
Plot a line graph on **income vs year** for the state of
**California**


#### Constraints

#### Test Cases
* Test Case 1
* Test Case 2
* Test Case 3

```python
import pandas as pd
df = pd.read_csv('DataFiles/Income.csv')
y = df.values[4]
x = df.columns
x
y
```

```python
years = []
for i in range(2, len(x)):
    years.append(x[i])
years
income = []
for i in range(2, len(y)):
    income.append(y[i])
income
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Income of California State')
plt.plot(years, income)
plt.grid(True)
plt.savefig('Figures/line-graph-income-California.png')
plt.show()
```

```python
width = 1/3
plt.title('Income of California')
plt.bar(years, income, width, color='Gray')
#plt.savefig('Figures/bar-graph-income-California.png')
```

### Problem 4 : 
#### Problem Statement
Generate a Scatter Plot for 50 random numbers

#### Constraints

#### Test Cases
* Test Case 1

```python
n = 50
x = np.random.rand(n) #Generate 50 random numbers
y = np.random.rand(n)#Generate 50 random numbers
colorlist = np.random.rand(n)#Generate 50 random color values
arealist = (15) #Generate 50 random areas of circles
plt.scatter(x, y, s=arealist, c=colorlist, alpha=10)
plt.savefig('Figures/scatter-plot-50-random-numbers.png')
```
